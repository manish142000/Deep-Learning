```python
#importing libraries
import numpy as np
from skimage.io import imread, imshow
from skimage.filters import prewitt_h, prewitt_v, sobel_h, sobel_v
```

```python
#reading the image_data
dog_image = imread('dog.jpg', as_gray=True)
#checking the shape
dog_image.shape
```

Extracting edges using prewitt filter

```python
prewitt_horizontal = prewitt_h(dog_image)
imshow(prewitt_horizontal, cmap = 'gray')
```

From above image we can see that horizontal lines are highlighted 

```python
prewitt_vertical = prewitt_v(dog_image)
imshow(prewitt_vertical, cmap = 'gray')
```

From the above image we can see that vertical edges are more highlighted

Extracting edges from images using Sobel Kernel

```python
Sobel_horizontal = sobel_h(dog_image)
imshow(Sobel_horizontal, cmap = 'gray')
```

```python
Sobel_vertical = sobel_v(dog_image)
imshow(Sobel_vertical, cmap = 'gray')
```
